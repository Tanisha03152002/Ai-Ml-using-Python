```python
from pydataset import data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = data('Pima.tr')
df.plot(kind="scatter",x="skin",y="bmi",alpha=.5,color='r')
plt.show()
```

```python
X_train , X_test, y_train , y_test = train_test_split(df.skin,df.bmi)
plt.scatter(X_train,y_train,label="TRAINING DATA",color='b',alpha=.5)
plt.scatter(X_test,y_test,label="TESTING DATA",color='g',alpha=.9)
plt.legend()
plt.title("Test Train Show")
plt.show()
```

```python
LR = LinearRegression()
LR.fit(X_train.values.reshape(-1,1),y_train.values)
y_pred = LR.predict(X_test.values.reshape(-1,1))
plt.scatter(X_test,y_test,label="ACTUAL TEST DATA",color='g',alpha=.8)
plt.plot(X_test,y_pred,label="LINEAR REGRESSION",color='b')
plt.legend()
plt.show()
```

```python
#Predict BMI Of Woman With Skin Fold 50
LR.predict(np.array([[50]]))[0]
```

```python
#Score this model
LR.score(X_test.values.reshape(-1,1),y_test.values)
```
